```python
import numpy as np
```

```python
number_list = []
number = 0
num = 0
while number<10**100:
    number = ((3+2*np.sqrt(2))**num-(3-2*np.sqrt(2))**num)**2
    number_list.append(number)
    num +=1
    
print('there are ',len(number_list),'numbers before 10**100')
print(number_list)
```

```python
number_list = [0]
number = 0
num = 0
while number<10**10**100:
    number = ((3+2*np.sqrt(2))**num-(3-2*np.sqrt(2))**num)**2
    number_list.append(number)
    num +=1
    
print('there are ',len(number_list),'numbers before 10**10**100')
```
