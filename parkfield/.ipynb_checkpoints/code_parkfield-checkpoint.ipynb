# Functions

```python
def build_filename(path,day):
    '''Cette focntion permet d'écrire automatiquement le chemin d'accès au fichier pour une date et un jour donné'''
    if day >= 100:
        filename = f'{path}/day_{day}.h5'
    elif day<100 and day >=10 :
        filename = f'{path}/day_0{day}.h5'
    elif day < 10:
        filename = f'{path}/day_00{day}.h5'
    return filename
    
    
def data_exist(file, station,loc):
    '''Cette fonction permet de déterminer si dans un fichier il y a bien des données concernant la station choisie
    La focntion prend en compte le fait qu'il puisse n'y avoir aucune donnée du network XN (donc aucune station du réseau n'enregistre ce jour là)
    Et ça vérifie la présence du station name au sein des keys du fichier'''
    exist = False
    #verifie que y'a bien des enregistrement dans 
    if list(file.keys()).count('XN')==1: #s'il  y a bien XN dans la liste des network du fichier
        if list(file['XN'].keys()).count(f'{station}.{loc}')>0:
            exist = True
    return exist

def get_trace(network,station,location,year,day, dayttype='raw',cmp='Z',processed=False, fs=10):
    '''Faire une fonction permettant de récup automatiquement une trace (qui build le nom, le path etc avec comme indication la trace le jour l'année etc )'''
    if processed==True:
        path = f'data_parkfield/data_{fs}.0hz_cooked/daily/glob/{year}'
    else:
        path = f'data_parkfield/data_{fs}.0hz/daily/glob/{year}'
    
    if dayttype == 'raw':
        filename = build_filename(path, day)
    else:
        day = day+175 #on rajoute 175 parce que ça commence à day 175 pas à day 0 
        if day > 365: #dans le cas où l'on indique comme jour le nombre de jour depuis le début de la mission si on est en 2002 alors on retire 365
            day = day-365 
        filename = build_filename(path, day)
    file = h5py.File(filename,'r')
    tr = file[f'{network}'][f'{station}.{location}'][cmp][:]
    return tr


def get_station_list(path):
    station_list = pd.read_csv(path, header=None, sep=' ', usecols=[8])
    station_list = list(station_list.values)
    station_list_clean = []
    for i in range(len(station_list)):
        station_list_clean.append(station_list[i][0])
    return station_list_clean


def date_to_day(starttime,endtime):
    '''Calcule le nombre de jours depuis le 24 juin ce qui sert ensuite à savoir dans quels fichier récupérer les données'''
    beginning = UTCDateTime('2001-06-24 00:00:00')
    daystart = int((starttime-beginning)/86400)   #il return la différence en seconde et on divise par 86400 pour avoir l'équivalent en jour 
    dayend = int((endtime-beginning)/86400)
    return daystart, dayend

def merge_traces(network,station,location,channel,starttime,endtime, processed=False,fs=10):
    '''Version généralisée qui permet de changer la date de début et de fin '''
    availability_matrix = np.loadtxt('availability_matrix') #il charge depuis le fichier, donc plus besoin de compiler la cellule du dessus
    station_list_clean = get_station_list('data_parkfield/stations.txt')
    index_station = station_list_clean.index(station)
        
    #setup les infos pour tr et tr full
    tr_full = obs.Trace()
    tr_full.stats.sampling_rate = fs
    tr_full.stats.network = network
    tr_full.stats.station = station
    tr_full.stats.location = '' #trcage parce qu'on use '00' pour le fichier sauf qu'en vrai c'est ''
    tr_full.stats.channel = channel
    tr = tr_full.copy()
    
    #on doit déterminer le jour de fichier de début et de fin + il faudra q
    daystart, dayend = date_to_day(starttime, endtime) #il return les days par rapport au dbéut enregistrement  donc faudra rajouter 175 pour le fichier à build
    #dans le cas où le daystart n'a pas de données, le risque est que ce ne soit pas interprété comme un gap par la fonction ppsd, il faut par conséquent recalculer le jour où commencer
    #et la date qui lui est alors associée
    if availability_matrix[index_station,daystart]==0:#si ça commence alors qu'il n'y a pas de données
        while availability_matrix[index_station,daystart]==0:
            daystart +=1 #on increase le jour de départ jusqu'à ne plus commencer par un gap
            starttime += 86400 #on recalcule alors la date de début en accord
    
    day_list_matrix = np.arange(daystart,dayend+1) #direct les indices à chercher dans l'availability matrix du coup puisque c'est par rapport à début enregistrement
    
    tr_full.stats.starttime = starttime #on peut maitnenant donnerun début à la trace full
    
    print('Loading data...')
    for day in tqdm(day_list_matrix):
        if availability_matrix[index_station,day]!=0: #donc s'il y a de la donnée ce jour là, focntionne pour 2001 comme 2002 , du coup si y'a plus de donnée la trace sera plus courte que prévu 
            tr.stats.starttime = starttime+(day-daystart)*86400 #il update le start time à chaque itération à partir du starttime donné initialement 
            #maintenant il faut ajuster le day pour chacune des années,puisque le day qu'on utilise c'est par rapport au daystart
            if day <= 190:
                year = 2001
                day_corr = day+175 #175 car on a des données à partir du 24 juin la date de début d'enregistrement
            else: #dans ce cas on est en 2002 donc retranche une année
                year = 2002
                day_corr = day+175-365 
            tr.data = np.float64(get_trace(network,station,location,year,day_corr,cmp=channel[2], processed=processed, fs=fs))
            tr_full += tr 
    return tr_full
    
def plot_psd(network,station,location,channel,starttime,endtime,outname='', processed=False, fs=10):
    tr = merge_traces(network,station,location,channel,starttime,endtime, processed=processed, fs=fs)
    #tr = tr_full
    st = obs.Stream()
    st.append(tr)
    #création d'une réponse intrumentale plate 
    paz = {'gain':1., 'sensitivity': 1., 'poles': [1-1j], 'zeros': [1-1j]}
    #et maintenant on use obspy
    print('Creating the PPSD...')
    ppsd = PPSD(tr.stats,metadata=paz, ppsd_length=3600, skip_on_gaps=True)  #ppsd sur une journée entière donc devrait faire 479 bins, sauf qu'une bonne partie du lot y'a pas de data =0
    ppsd.add(tr)
    print('Plotting the PPSD...')
    if outname == '':
        ppsd.plot()
    else:
        ppsd.plot(filename=outname)
        
        
def make_gif(station,duration):
    frames = [Image.open(image) for image in glob.glob(f"PSDS/{station}/*.png")]
    frame_one = frames[0]
    frame_one.save(f"PSDS/{station}/animation.gif", format="GIF", append_images=frames,
               save_all=True, duration=duration, loop=0)
    
# 3 fonctions en dessous servent à trouver les dates qui posent problème, qui peuvent alors être comparées avec les périodes où les PSD font n'imp 
    
def convert_problem_list(problem_list):
    '''Cette focntion prend en entrée la liste de 0 et de 1 qui dit si chaque jour depuis le 24 juin (index 0) il y a eu un problème et il return la liste des dates où il y a eu des problèmes'''
    date = UTCDateTime('2001-06-24 00:00:00')
    problematic_dates = []
    for day in problem_list:
        if day==1: #s'il y a un problème
            problematic_dates.append(date) #alors on rajoute la date à la liste des dates problématiques
        date += 86400 #à chaque itération il rajoute 1 jour
    return problematic_dates

def find_problem(network,station,location,channel,starttime,endtime):
    '''La problem liste peut prendre 3 valeurs : -1 si pas de donnée, 0 si pas de problème et 1 si problème de saturation'''
    daystart, dayend = date_to_day(starttime,endtime) #on sait le nb de jour depuis le 24 juin
    
    day_list = np.arange(daystart,dayend+1) 

    station_list_clean = get_station_list('data_parkfield/stations.txt')
    index_station = station_list_clean.index(station) # on trouve 'lindex de notre station dans la matrice
    availability_matrix = np.loadtxt('availability_matrix') #import de la matrice      
    
    problem_list = np.zeros(len(day_list))
    
    print('Looking for troubles, please wait...')
    for i in tqdm(range(len(problem_list))):
        if availability_matrix[index_station,day_list[i]]==1: #donc si y'a de la donnée 
            if day_list[i] <= 190: #on est en 2001
                year_day = day_list[i]+175
                problem_list[i] = find_daily_problem(network,station,location,channel,2001,year_day)
            else: #2002
                year_day = day_list[i]+175-365
                problem_list[i] = find_daily_problem(network,station,location,channel,2002,year_day)
        else:
            problem_list[i] = -1 #on met -1 pour que la fonction convert sache qu'elle ne doit pas interpréter ces dates
    problem_list_converted = convert_problem_list(problem_list)
    return problem_list_converted #il return une liste de dates où y'a problème; on pourra aussi lui faire output les jours de l'année où il  ya pb qui pourrait être used pour le beam!
    
def find_daily_problem(network,station,location,channel,year,day):
    '''Cette fonction permet de trouver s'il y a un problème de la saturation à un jour donné'''
    problem = 0
    trace = np.abs(get_trace(network,station,location,year,day)) #trace en valeur absolue
    enveloppe = np.median(trace) #on cherche la valeur médiane de la trace, donc pas affectée par des moments où il y a de la saturation
    for i in range(len(trace)):
        if trace[i] > 1e4*enveloppe: #si c'est nettement plus fort que l'enveloppe c'est que ça sature
            problem = 1
    return problem



############### POUR LE BEAM ###################

def extract_coordinates(path,sta_name=False):
    '''Fonction qui récupère en input la liste des station et qui va  extraire les coordonnées
    L'option sta_name permet de récupérer les noms des stations en plus des autres listes'''
    #laoding the station_list, cette version de station list devra être la version avec les stations non utilisées virées 
    latitudes = pd.read_csv(path, header=None, sep='    ',usecols=[4],engine='python').values
    longitudes = pd.read_csv(path, header=None, sep='    ', usecols=[5],engine='python').values
    heights = pd.read_csv(path, header=None, sep='    ', usecols=[6], engine='python').values
    latitudes_clean = []
    longitudes_clean = []
    heights_clean = []
    if sta_name==True:
        networks = pd.read_csv(path, header=None, sep='    ',usecols=[1],engine='python').values
        networks_clean = []
        stations = pd.read_csv(path, header=None, sep='    ',usecols=[2],engine='python').values
        stations_clean = []

    for i in range(len(latitudes)):
        latitudes_clean.append(latitudes[i][0])
        longitudes_clean.append(longitudes[i][0])
        heights_clean.append(heights[i][0])
        if sta_name==True:
            networks_clean.append(networks[i][0])
            stations_clean.append(stations[i][0])
        
    latitudes_clean = np.array(latitudes_clean)    
    longitudes_clean = np.array(longitudes_clean)
    heights_clean = np.array(heights_clean)
    
    if sta_name == True:
        networks_clean = np.array(networks_clean)
        stations_clean = np.array(stations_clean)
        return longitudes_clean,latitudes_clean,heights_clean, networks_clean, stations_clean 
    else:
        return longitudes_clean,latitudes_clean,heights_clean

def compute_barycenter(path):
    '''Fonction qui récupère en input la liste des station et qui va calculer les coordonnées du barycentre de l'array'''
    longitudes_clean,latitudes_clean,heights_clean = extract_coordinates(path)
    #loading the transformers to convert coordinates to     
    trans_GPS_to_XYZ = Transformer.from_crs(4979, 4978, always_xy=True)
    trans_XYZ_to_GPS = Transformer.from_crs(4978,4979,always_xy=True)
    #on a créé les convertisseurs    
    X,Y,Z = trans_GPS_to_XYZ.transform(longitudes_clean, latitudes_clean, heights_clean)
    X_bary = np.mean(X) 
    Y_bary = np.mean(Y)
    Z_bary = np.mean(Z)
    lon_bary,lat_bary,height_bary = trans_XYZ_to_GPS.transform(X_bary,Y_bary,Z_bary)
    return lon_bary,lat_bary,height_bary 
```

# Creating the availability matrix

```python
#on peut faire une matrUTCDateTime en bouclant sur  les années et sur les jours des années et sur les stations, e    ce qu'on peut faire c'est qu'on peut rajouter 365 aux values de la deuxième 
#année tout simple et on boucle sur les jours    et boucle sur les stations   et on met 1 si y'a de la donnée et on met 0 si y'a pas de donnée 

path2001 = "data_parkfield/data_10.0hz/daily/glob/2001/"
path2002 = "data_parkfield/data_10.0hz/daily/glob/2002/"

#va falloir faire la liste des stations à patir du stations.txt
station_list_clean = get_station_list('data_parkfield/stations.txt')
#c'est bon on a une station list clean maintenant

#maintenant on doit s'occuper des jours 

years_list = [2001,2002]

daymin2001 = 175   #24 juin 2001
daymax2001 = 365
daymin2002 = 1
daymax2002 = 288 # 15 octobre 2002  

day_list2001 = np.arange(daymin2001,daymax2001+1)
day_list2002 = np.arange(daymin2002,daymax2002+1)

day_list = np.array((list(day_list2001)+list(day_list2002))) #donc ça recommence à 1 après 365, et on connait la durée dans chaque année

#on peut maintenant créer la matrice qui va nous renseigner s'il y a des données où non à la station pour le jour et la station donnée

data_matrix = np.zeros((len(station_list_clean), len(day_list)))
```

```python
#maintenant qu'on sait où l'on doit ajouter de la donnée il faut que l'on effectue une boucle sur l'intégralité des jours et sur les différentes stations 
#cette boucle dans prendre en compte le switch d'année 2001 à 2002 

day_list_plot = np.arange(len(day_list))

for year in years_list: #on boucle sur les années 
    if year==2001:
        for day in day_list2001:
            for i in range(len(station_list_clean)):
                filename = build_filename(path2001,day)
                file = h5py.File(filename,'r') 
                exist = data_exist(file, station_list_clean[i], '00')
                if exist==True: 
                    data_matrix[i,day-daymin2001] = 1  #on retire daymin2001 pour que ça corresponde au bon index
            
    elif year==2002:
        for day in day_list2002:
            for i in range(len(station_list_clean)):
                filename = build_filename(path2002,day) #on doit retirer 365 sinon il va chercher des fichiers avec des jours trop grands
                file  = h5py.File(filename,'r')
                exist = data_exist(file, station_list_clean[i], '00')
                if exist==True:
                    data_matrix[i,day-day_list2002[0]+len(day_list2001)] = 1  #retire le daymin2 (celui de la liste car c'est celui increase de 365) pour que 1 soit le 0, mais rajoute la len de l'année 2001 
         
    
plt.figure(figsize=(10,10))
plt.pcolormesh(day_list_plot,station_list_clean,data_matrix)
plt.xlabel('Day since the installation of the first stations')
plt.title('Existence of data of the different stations throughout the mission')
plt.savefig('data_availability.png', bbox_inches='tight')

np.savetxt('availability_matrix',data_matrix) #on export l'availaibility matrix, cette matrice nous servira par la suite à savoir sur quelles stations et jours boucler pour faire anlyse des PSD
```

```python
#l'objectif est de retirer les 
daymin2001 = 175   #24 juin 2001
daymax2001 = 365
daymin2002 = 1
daymax2002 = 288 # 15 octobre 2002  
day_list2001 = np.arange(daymin2001,daymax2001+1)
day_list2002 = np.arange(daymin2002,daymax2002+1)
day_list = np.array((list(day_list2001)+list(day_list2002))) #donc ça recommence à 1 après 365, et on connait la durée dans chaque année
day_list_plot = np.arange(len(day_list))
station_list_clean = get_station_list('data_parkfield/stations.txt')

new_data_matrix = np.loadtxt('availability_matrix')

bad_stations = ['BUZZ','CRAB','MRED','VINE']

bad_times = [[ ], #pour BUZZ
            [], #pour CRAB
            [], #pour MRED
            []] #pour BUZZ

bad_times[0] = [(date_to_day(UTCDateTime('2001-12-21'),UTCDateTime('2001-12-30'))), (date_to_day(UTCDateTime('2002-01-08'),UTCDateTime('2002-03-15'))), (date_to_day(UTCDateTime('2002-04-22'),UTCDateTime('2002-05-02')))]
bad_times[1] = [(date_to_day(UTCDateTime('2002-01-11'),UTCDateTime('2002-03-15')))]
bad_times[2] = [(date_to_day(UTCDateTime('2001-11-25'),UTCDateTime('2001-11-26'))),(date_to_day(UTCDateTime('2001-11-30'),UTCDateTime('2001-12-08'))),(date_to_day(UTCDateTime('2001-12-25'),UTCDateTime('2001-12-28'))),(date_to_day(UTCDateTime('2002-01-08'),UTCDateTime('2002-03-15'))),(date_to_day(UTCDateTime('2002-05-05'),UTCDateTime('2002-05-09'))),(date_to_day(UTCDateTime('2002-05-17'),UTCDateTime('2002-05-19'))),(date_to_day(UTCDateTime('2002-08-23'),UTCDateTime('2002-10-08')))]
bad_times[3] = [(date_to_day(UTCDateTime('2001-10-02'),UTCDateTime('2001-10-11'))),(date_to_day(UTCDateTime('2001-11-23'),UTCDateTime('2001-11-24'))),(date_to_day(UTCDateTime('2001-11-27'),UTCDateTime('2001-11-29'))),(date_to_day(UTCDateTime('2002-01-08'),UTCDateTime('2002-01-18'))),(date_to_day(UTCDateTime('2002-02-06'),UTCDateTime('2002-03-13'))),(date_to_day(UTCDateTime('2002-03-30'),UTCDateTime('2002-04-01')))]

#print(bad_times)


for i in range(len(bad_stations)):
    ind = station_list_clean.index(bad_stations[i]) #on trouve l'index de la mauvaise station dans l'availability matrix 
    
    for j in range(len(new_data_matrix[0,:])): #on recolore aussi la trace pour lui donner une couleur un peu différente, signifiant que on l'a dump
        #ça permet surtout pour la fonction pour retirer les mauvais jours du beam, de savoir que malgré que y'a de la bonne data sur ces stations, 
        #elles ont été dumped!
        if new_data_matrix[ind,j] == 1: #si y'a de la donnée ce jour là
            new_data_matrix[ind,j] = 0.5 #alors on remplace par 0.5 pour dire que on tej cette trace
        
    
    for j in range(len(bad_times[i])): #on cherche l'ensemble des mauvaises périodes
        bad_days = np.arange(bad_times[i][j][0],bad_times[i][j][1]+1) #il fait un array contenant les indices de l'availability matrix devant être remplacés  
        for day in bad_days:
            if new_data_matrix[ind,day] == 0.5: #rajoute cette condition pour qu'il mette pas en rouge si c'est 0!!
                new_data_matrix[ind,day] = -1 
        
np.savetxt('availability_matrix_nobad', new_data_matrix)
        
cmap = 'RdYlGn'     
plt.figure(figsize=(10,10))
plt.pcolormesh(day_list_plot,station_list_clean,new_data_matrix, cmap=cmap)
plt.xlabel('Day since the installation of the last stations')
#plt.title('Existence of data of the different stations throughout the mission')
#plt.colorbar()
plt.savefig('data_availability_with_bad.pdf', bbox_inches='tight')
```

```python
#on importe les failles depuis le fichier kmz 
import zipfile
import geopandas as gpd
from bs4 import BeautifulSoup
from shapely.geometry import LineString

def extract_kmz_data(kmz_file,extent=None):
    with zipfile.ZipFile(kmz_file, 'r') as kmz:
        kml_file = kmz.extract('doc.kml', './')
    
    with open(kml_file, 'r') as kml:
        soup = BeautifulSoup(kml, 'xml')
        placemarks = soup.find_all('Placemark')
        
        paths = []
        for placemark in placemarks:
            name = placemark.find('name').text
            coordinates = placemark.find('coordinates').text.strip()
            path = [tuple(map(float, coord.split(','))) for coord in coordinates.split()]
            if extent!=None:
                for i,coords in enumerate(path): #check toutes les coordonnées du path
                    if extent[0]<=coords[0]<=extent[1] and extent[2]<=coords[1]<=extent[3]: #di y'a un seul point dans la range 
                        paths.append((name, path)) #alors on add 
                        break #on break pour passer au path suivant au lieu de continuer d'itérer dans le même path car déjà ajouté aux paths
            else:
                paths.append((name, path))
    
    return paths

def plot_paths_on_map(ax, paths_data):
    gdfs = []
    for name, path in paths_data:
        # Convert list of tuples to LineString geometry
        line_string = LineString(path)
        gdf = gpd.GeoDataFrame({'Name': [name], 'geometry': [line_string]}, crs='EPSG:4326')
        gdfs.append(gdf)
    combined_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True), crs='EPSG:4326')
    combined_gdf.plot(ax=ax, color='red', zorder=3, linewidth=0.4)

# extent = [-122.1,-119.5,34.5,37.5]   #pour la full map
extent = [-121.3,-120.4,35.5,37]
fault_data = extract_kmz_data('faults_full.kmz', extent=extent)
print('faults successfully imported')
```

```python
def concat_traces(network, station, location):
    '''Cette fonction permet de concaténer les traces de la station sélectionnée pour en faire une seule même trace'''
    trace_full = []
    exist_full = []
    availability_matrix = np.loadtxt('availability_matrix') #il charge depuis le fichier, donc plus besoin de compiler la cellule du dessus
    index_station = station_list_clean.index(station) #il donne l'indice dans la matrice correspondant à la station
    for year in years_list:
        if year == 2001:
            print('Loading 2001...')
            for day in tqdm(day_list2001):
                if availability_matrix[index_station,day-day_list2001[0]]==1: #donc s'il y a de la donnée ce jour là  
                    trace = list(get_trace(network,station,location,year,day))
                    exist_full +=1
                else:
                    trace = list(np.zeros(864000)) #on met 864000 zéros pour les jours où y'a pas de donnée
                    exist_full += 0
                trace_full += trace #dans les deux cas on concatène  
                
        elif year == 2002:
            print('Loading 2002...')
            for day in tqdm(day_list2002):
                if availability_matrix[index_station, day-day_list2002[0]+len(day_list2001)]==1:
                    trace = list(get_trace(network,station,location,year,day))
                    exist_full +=1
                else:
                    trace = list(np.zeros(864000))
                    exist_full +=0
                trace_full += trace
                
    #on a maintenant concaténé les diverses traces de la station on les retransforme en np array pour que ce soit plus adapté
    trace_full = np.array(trace_full)
    return trace_full, exist_full
    
```

```python
def merge_traces(network,station,location,channel):
    fs = 10
    availability_matrix = np.loadtxt('availability_matrix') #il charge depuis le fichier, donc plus besoin de compiler la cellule du dessus
    index_station = station_list_clean.index(station)
    
    starttime = UTCDateTime('2001-06-24 00:00:00')
    for day in day_list2001:
        if availability_matrix[index_station,day-day_list2001[0]]==1:
            startday = day
    starttime += (startday-day_list2001[0])*86400 #c'est bon on a maintenant notre véritable startime, là où on commence varitablement à record
    
    
    #setup les infos pour tr et tr full
    tr_full = obs.Trace()
    tr_full.stats.sampling_rate = fs
    tr_full.stats.starttime = starttime
    tr_full.stats.network = network
    tr_full.stats.station = station
    tr_full.stats.location = '' #trcage parce qu'on use '00' pour le fichier sauf qu'en vrai c'est ''
    tr_full.stats.channel = channel
    tr = tr_full.copy()
    
    
    for year in years_list:
        if year == 2001:
            print('Loading 2001...')
            for day in tqdm(day_list2001):
                if availability_matrix[index_station,day-day_list2001[0]]==1: #donc s'il y a de la donnée ce jour là  
                    tr.stats.starttime = starttime+(day-startday)*86400 #il update le start time en fonction du jour    et le startday c'est le jour où commence vraiment à enregistrer!
                    tr.data = np.float64(get_trace(network,station,location,year,day))
                    tr_full += tr 
        elif year == 2002:
            print('Loading 2002...')
            for day in tqdm(day_list2002):
                if availability_matrix[index_station, day-day_list2002[0]+len(day_list2001)]==1:
                    tr.stats.starttime = UTCDateTime('2002-01-01 00:00:00')+(day-day_list2002[0])*86400 #il update le start time en fonction du jour
                    tr.data = np.float64(get_trace(network,station,location,year,day))
                    tr_full += tr
    return tr_full
    
def plot_psd(network,station,location,channel,outname=''):
    tr = merge_traces(network,station,location,channel)
    #tr = tr_full
    st = obs.Stream()
    st.append(tr)
    #création d'une réponse intrumentale plate 
    paz = {'gain':1., 'sensitivity': 1., 'poles': [1-1j], 'zeros': [1-1j]}
    #et maintenant on use obspy
    print('Creating the PPSD...')
    ppsd = PPSD(tr.stats,metadata=paz, ppsd_length=86400, skip_on_gaps=True)  #ppsd sur une journée entière donc devrait faire 479 bins, sauf qu'une bonne partie du lot y'a pas de data =0
    ppsd.add(tr)
    print('Plotting the PPSD...')
    if outname == '':
        ppsd.plot()
    else:
        ppsd.plot(filename=outname)
```
